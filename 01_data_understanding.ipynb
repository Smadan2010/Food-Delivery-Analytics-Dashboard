# 01 - Data Understanding


```python
# Import pandas library
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
# Load the dataset
df = pd.read_csv('ONINE_FOOD_DELIVERY_ANALYSIS.csv')

print("Dataset loaded successfully!")
print("\n")
```

## Step 3: Check the Shape

```python
# Check the shape
print("Shape of the dataset:")
print(df.shape)
print("\n")
```

## Step 4: View First Few Rows

```python
# Display first 5 rows
print("First 5 rows of the dataset:")
print("\n")
df.head()
```

## Step 5: Check Column Names


```python
# Display all column names
print("Column names in the dataset:")
print("\n")
for i, col in enumerate(df.columns, 1):
    print(f"{i}. {col}")
```

## Step 6: Check Data Types


```python
# Check data types
print("Data types of each column:")
print("\n")
print(df.dtypes)
print("\n")

# Count each data type
print("Summary of data types:")
print(df.dtypes.value_counts())
```

## Step 7: Check Missing Values

```python
# Check for missing values
print("Missing values in each column:")
print("\n")
missing_values = df.isnull().sum()
print(missing_values)
print("\n")

# Show percentage of missing values
print("Percentage of missing values:")
print("\n")
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_data = pd.DataFrame({
    'Missing_Count': missing_values,
    'Percentage': missing_percentage
})
print(missing_data[missing_data['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False))
```

## Step 8: Get Summary Statistics


```python
# Get summary statistics for numeric columns
print("Summary statistics for numeric columns:")
print("\n")
df.describe()
```

## Step 9: Check Unique Values in Categorical Columns

```python
# List of categorical columns to check
categorical_columns = ['City', 'Customer_Gender', 'Cuisine_Type', 'Payment_Mode', 
                       'Order_Status', 'Order_Day']

# Check unique values in each categorical column
print("Unique values in categorical columns:")
print("\n")

for col in categorical_columns:
    if col in df.columns:
        print(f"{col}:")
        print(f"  Number of unique values: {df[col].nunique()}")
        print(f"  Values: {df[col].unique()}")
        print("\n")
```

## Step 10: Check for Duplicate Rows

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

if duplicates > 0:
    print("\nWe have duplicate rows. We will handle them in the cleaning step.")
else:
    print("\nNo duplicate rows found. Good!")
```
